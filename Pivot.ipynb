```python
import timeit
start = timeit.default_timer()
newdf = pd.DataFrame(sorted(df2['Acct_Num'].unique()), index = sorted(df2['Acct_Num'].unique()), columns = ['Acct_Num'])
for j in sorted(df2['Prod_Num'].unique()):
    newdf[('Prod_' + str(j) + '_Qty')] = 0
    newdf[('Prod_' + str(j) + '_Vol')] = 0
for i in sorted(newdf['Acct_Num']):
    for j in sorted(df2['Prod_Num'].unique()):
        if df2[(df2['Acct_Num'] == i) & (df2['Prod_Num'] == j)].groupby(['Acct_Num', 'Prod_Num']).sum().shape[0] == 0:
            newdf.set_value(i, ('Prod_' + str(j) + '_Qty'), 0)
            newdf.set_value(i, ('Prod_' + str(j) + '_Vol'), 0)
        else:    
            a = df2[(df2['Acct_Num'] == i) & (df2['Prod_Num'] == j)].groupby(['Acct_Num', 'Prod_Num']).sum().iloc[0,0]
            b = df2[(df2['Acct_Num'] == i) & (df2['Prod_Num'] == j)].groupby(['Acct_Num', 'Prod_Num']).sum().iloc[0,1]
            newdf.set_value(i, ('Prod_' + str(j) + '_Qty'), a)
            newdf.set_value(i, ('Prod_' + str(j) + '_Vol'), b)
stop = timeit.default_timer()
print('Time: ', stop - start)  
```
